```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
#Active Cases

covid_df['Active_Cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.head()
```

```python
covid_df
```

```python
# create a dataframe "statewise" which is the pivot_table of dataframe "covid_df". 
# which show the max values of 'Confirmed','Cured' and 'Deaths' Columns


statewise=pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc="max")
statewise
```

```python
"""create a dataframe "statewise" which is the pivot_table of dataframe "covid_df". 
which show the 'total sum'  of 'Confirmed','Cured' and 'Deaths' Columns. """



statewise=pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc="sum")
statewise
```

```python
# add 'Recovery Rate' column at "statewise" dataframe columns 

statewise["Recovery Rate"]=statewise["Cured"]*100/statewise["Confirmed"]
statewise
```

```python
# add 'Mortality Rate' column at "statewise" dataframe columns 


statewise["Mortality Rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
statewise
```

```python
# sort the data in decreasing order of 'Confirmed' column 


statewise = statewise.sort_values(by="Confirmed",ascending=False)
statewise
```

```python
# top 10 state  based on active  cases

top_10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']]\
.sort_values(by=['Active_Cases'],ascending=False).reset_index()
top_10_active_cases.head(10)
```

```python
#another way
covid_df.groupby('State/UnionTerritory').max()[['Active_Cases','Date']]\
.sort_values('Active_Cases',ascending=False).reset_index().head(10)
```

```python
#plot top 10 states in visualization manner
top_10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']]\
.sort_values(by=['Active_Cases'],ascending=False).reset_index()

fig=plt.figure(figsize=(12,5))

sns.barplot(data=top_10_active_cases.iloc[:10],y='Active_Cases',x='State/UnionTerritory',linewidth=2,edgecolor='red')

plt.title("top 10 states with most active cases")

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()


```

```python
#top 10 states on total number of death reported
top_10_deaths_cases=covid_df.groupby('State/UnionTerritory').max()[['Deaths','Date']]\
.sort_values('Deaths',ascending=False).reset_index()

fig=plt.figure(figsize=(16,9))
sns.barplot(data=top_10_deaths_cases.iloc[:10],x='State/UnionTerritory',y='Deaths',linewidth=2,edgecolor='orange')

plt.title("top 10 states with most deaths ",size=25)
plt.xlabel('States')
plt.ylabel('Total Death Cases')
plt.show()

# top_10_deaths_cases

```

```python
# drop some missinng values columns
vaccination=vaccine_df.drop(columns=['Sputnik V (Doses Administered)','AEFI','18-44 Years (Doses Administered)'
                                     ,'45-60 Years (Doses Administered)','60+ Years (Doses Administered)'],axis=1)
vaccination.head(10)
```

```python
#to create a piplot to see the vaccineotion of male and female

male=vaccination["Male(Individuals Vaccinated)"].sum()
female=vaccination["Female(Individuals Vaccinated)"].sum()
px.pie(names=["Male","Female"],values=[male,female],title="male and female Vaccinaion")
```

```python
#remove all the states where state=India

vaccine=vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
#rename last column "Total Individuals Vaccinated" to  just 'Total'

vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
vaccine
```

```python
#states with most vaccinated individual

max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python
#now we plot above data visualizatioon

fig=plt.figure(figsize=(10,5))
plt.title("top 5 vaccinated States in India",size=20)
x=sns.barplot(data=max_vac,y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

```python
#top 5 states with least vaccinated individual
min_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac=max_vac.sort_values('Total',ascending=True)[:5]
min_vac
```

```python
#now we plot above data visualizatioon

fig=plt.figure(figsize=(10,5))
plt.title("top 5 min  vaccinated States in India",size=20)
x=sns.barplot(data=min_vac,y=min_vac.Total,x=min_vac.index,linewidth=2,edgecolor='black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
